```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
data = np.random.randn(10000)

plt.hist(data, bins=50)

plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")

plt.show()
```

```python
x = ['A', 'B', 'C', 'D', 'E']
y = np.random.randint(1, 10, 5)

plt.bar(x, y)

plt.title("Column Chart")
plt.xlabel("Categories")
plt.ylabel("Values")

plt.show()
```

```python
sizes = [20, 25, 21, 10, 8, 37]

plt.pie(sizes)

plt.title("Pie Chart")

plt.show()
```

```python
x = np.arange(-10, 10, 0.5)
y = np.arange(-8, 9, 0.5)
x, y = np.meshgrid(x, y)
r = np.sqrt(x**2 + y**2)
z = np.sin(r)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(x, y, z)

ax.set_title("3D Graphics")
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_zlabel("Z axis")
plt.show()
```

```python
x = np.random.rand(30)
y = np.random.rand(30)
colors = np.random.rand(30)
sizes = np.random.randint(30, 130, 30)

plt.scatter(x, y, c=colors, s=sizes)

plt.title("Point Chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

```python
x = np.arange(0, 50, 0.4)
y = np.sin(x)

plt.plot(x, y)

plt.title("Sinüs Grafiği")
plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")

plt.show()
```
