```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import svm
```

```python
df = pd.read_csv('/content/diabetes.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df['Outcome'].value_counts()
```

```python
df.groupby('Outcome').mean()
```

```python
X = df.drop(columns = 'Outcome', axis = 1)
Y = df['Outcome']
```

```python
X
```

```python
Y
```

```python
scaler = StandardScaler()
```

```python
standarized_data = scaler.fit_transform(X)
```

```python
standarized_data
```

```python
X = standarized_data
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
classifier = svm.SVC(kernel = 'linear')
```

```python
classifier.fit(X_train, Y_train)
```

```python
X_train_pred = classifier.predict(X_train)
```

```python
accuracy_training = accuracy_score(X_train_pred, Y_train)
```

```python
print('accuracy of training part :', accuracy_training)
```

```python
X_test_pred = classifier.predict(X_test)
accuracy_test = accuracy_score(X_test_pred, Y_test)
```

```python
print('accuracy of test part : ', accuracy_test)
```

```python
#we are making new prediction, we onverting the new data to numpy array, reshape, and transforming as trained data
new_data = (7, 196, 90, 0, 0, 39.8, 0.451, 41)
new_data_into_numpy = np.asarray(new_data)
new_data_reshaped = new_data_into_numpy.reshape(1, -1)
new_data_standarized = scaler.transform(new_data_reshaped)

print(new_data_standarized)
```

```python
prediction = classifier.predict(new_data_standarized)

prediction
```

```python
if(prediction[0] == 1):
  print("this person is diabetic")
else:
  print("this person is not diabetic")
```
